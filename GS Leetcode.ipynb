```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){
    

    int n;
    cin>>n;
    vector<int>arr(n);
    for(int i=0;i<n;i++){
        cin>>arr[i];
    }

    int largest=INT_MIN;
    int secondlargest=INT_MIN;

    for(int i=0;i<n;i++){
        if(arr[i]>largest){
            secondlargest=largest;
            largest=arr[i];
        }else if(arr[i]>secondlargest and arr[i]<largest){
            secondlargest=arr[i];
        }

    }
    cout<<largest<<endl;
    cout<<secondlargest<<endl;



    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);    
    // int _t;cin>>_t;while(_t--)
    solve();
}

# First Unique Character in a String

```cpp
class Solution {
public:
    int firstUniqChar(string s) {
        unordered_map<char,int>mp;

        for(int i=0;i<s.size();i++){
            mp[s[i]]++;
        }
        int ans=-1;

        for(int i=0;i<s.size();i++){
            if(mp[s[i]]==1){
                ans=i;
                break;
            }
        }
        return ans;
    }
};

# Given string str of size N containing only lowercase English letters. The task is to encrypt the string such that the substrings having same prefix as itself are replaced by a *. Generate the encrypted string. Input: str = “ababcababcd”
# Output: ab*c*d
# Explanation: The substring “ababc” starting from 5th index (0 based indexing) can be replaced by a ‘*’. So the string becomes “ababcababcd” ->  “ababc*d”. Now the substring “ab” starting from 2nd index can again be replaced with a ‘*’. So the string becomes “ab*c*d”

```cpp

#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){
    

    string s;
    cin>>s;
    int n=s.size();
    int mid=n/2;
    stack<string>st;
    while(mid>0){

        int i=0,j=mid;
        while(i<mid && j<n && s[i]==s[j]){
            i++;
            j++;
        }
        if(i==mid){
            st.push(s.substr(j));
            s=s.substr(0,mid);
            mid=s.size();
        }else{
            mid--;
        }

    }


    while(!st.empty()){
        s.push_back('*');
        s+=st.top();
        st.pop();
    }
    cout<<s<<endl;
    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);    
    // int _t;cin>>_t;while(_t--)
    solve();
}

```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/



struct color{
    string name;
    int r,g,b;
};


int binaryStringtoint(string str){
    int value=0;
    for(auto ch:str){
        value=(value<<1)+(ch-'0');
    }
    return value;
}
void solve(){

    string s;
    cin>>s;

    vector<color>pure={
        {"Black",0,0,0},
        {"White",255,255,255},
        {"Red",255,0,0},
        {"Green",0,255,0},
        {"Blue",0,0,255}
    };


    int r=binaryStringtoint(s.substr(0,8));
    int g=binaryStringtoint(s.substr(8,8));
    int b=binaryStringtoint(s.substr(16,8));


    int ans=INT_MAX;

    vector<string>closest;
    for(auto color:pure){
        int dr=r-color.r;
        int dg=g-color.g;
        int db=b-color.b;
        int dis=dr*dr+dg*dg+db*db;
        if(ans>dis){
            ans=dis;
            closest.clear();
            closest.push_back(color.name);
        }else if(dis==ans){
            closest.push_back(color.name);
        }
    }
    if(closest.size()==1){
        cout<<closest[0]<<endl;
    }else{
        cout<<"Ambiguous"<<endl;
    }


    

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);    
    // int _t;cin>>_t;while(_t--)
    solve();
}

```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){

	
	
	vector<int>arr={2,1,4,3};
	int n=arr.size();
	long long f=0;
	long long s=0;

	bool isf=true;
	bool isR=false;


	deque<int>dq(arr.begin(),arr.end());
	for(int round=0;round<n;round++){
		if(dq.empty()){
			break;
		}


		int current;
		if(!isR){
			current=dq.front();
			dq.pop_front();
		}else{
			current=dq.back();
			dq.pop_back();
		}


		if(isf){
			f+=current;
		}else{
			s+=current;
		}

		if(current%2==0){
			isR=!isR;
		}
		isf=!isf;
	}

	cout<<f-s<<endl;
    

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}
