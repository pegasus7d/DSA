```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){

    int n;
    cin>>n;
    vector<int>tag(n);
    for(int i=0;i<n;i++){
        cin>>tag[i];
    }

    int q;
    cin>>q;
    while(q--){
        int l,r;
        cin>>l>>r;
        l--;
        r--;

        map<int,int>mp;
        for(int i=l;i<=r;i++){
            mp[tag[i]]++;
        }
        int cnt=0;
        for(auto &it:mp){
            cnt+=((it.second)/2);
        }
        cout<<cnt<<endl;

    }


    

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);    
    // int _t;cin>>_t;while(_t--)
    solve();
}

```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){

    string s="abcde";
    // cin>>s;
    int a=2,b=4,k=26;
    // cin>>a>>b>>k;


   
    unordered_map<char,int>mp;
    unordered_map<string,int>freq;
    
    int i=0,j=0;
    while(j<s.size()){
        mp[s[j]]++;

        if(mp.size()<=k){

            for(int len=a;len<=b && j-i+1>=len;len++){
                freq[s.substr(j+1-len,len)]++;
            }
            


        }else if(mp.size()>k){
            while(mp.size()>k){

                mp[s[i]]--;
                if(mp[s[i]]==0){
                    mp.erase(s[i]);
                }
                i++;
                
            }

            for(int len=a;len<=b && j-i+1>=len;len++){
                freq[s.substr(j+1-len,len)]++;
            }
        }

        j++;
    }


    int ans=0;
    for (auto it:freq) {
        ans=max(ans,it.second);
    }
    cout<<ans<<endl;


    

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);    
    // int _t;cin>>_t;while(_t--)
    solve();
}  

```cpp
class Solution {
public:
    int maximizeGreatness(vector<int>& nums) {
        int n=nums.size();
        sort(nums.begin(),nums.end());
        int ans=0;
        int i=0,j=1;

        while(j<n){
            if(nums[i]<nums[j]){
                ans++;
                i++;
                j++;
            }else{
                j++;
            }
        }
        return ans;
    }

};

```cpp
// CPP program to find longest even sum 
// subsequence. 
#include <bits/stdc++.h> 
using namespace std; 

// Returns sum of maximum even sum subsequence 
int maxEvenSum(int arr[], int n) 
{ 
	// Find sum of positive numbers 
	int pos_sum = 0; 
	for (int i = 0; i < n; ++i) 
		if (arr[i] > 0) 
			pos_sum += arr[i]; 

	// If sum is even, it is our 
	// answer 
	if (pos_sum % 2 == 0) 
		return pos_sum; 

	// Traverse the array to find the 
	// maximum sum by adding a positive 
	// odd or subtracting a negative odd 
	int ans = INT_MIN; 
	for (int i = 0; i < n; ++i) { 
		if (arr[i] % 2 != 0) { 
			if (arr[i] > 0) 
				ans = max(ans, pos_sum - arr[i]); 
			else
				ans = max(ans, pos_sum + arr[i]); 
		} 
	} 

	return ans; 
} 

// driver program 
int main() 
{ 
	int a[] = { -2, 2, -3, 1 }; 
	int n = sizeof(a) / sizeof(a[0]); 
	cout << maxEvenSum(a, n); 
	return 0; 
} 


```cpp
#include <iostream>
#include <vector>
#include <algorithm>

using namespace std;

// Helper function to check if we can achieve target throughput with the given budget
bool canAchieveThroughput(const vector<int>& throughput, const vector<int>& scaling_cost, int budget, int target_throughput) {
    int n = throughput.size();
    int total_cost = 0;

    
    for (int i = 0; i < n; ++i) {
        if (throughput[i] >= target_throughput) {
            
            continue;
        }
        
       
        int required_scale = (target_throughput + throughput[i] - 1) / throughput[i] - 1;
        int cost_for_service = required_scale * scaling_cost[i];
        
        
        total_cost += cost_for_service;
        
        
        if (total_cost > budget) return false;
    }

   
    return total_cost <= budget;
}

int maxPipelineThroughput(const vector<int>& throughput, const vector<int>& scaling_cost, int budget) {
    int left = *min_element(throughput.begin(), throughput.end());  
    int right = *max_element(throughput.begin(), throughput.end()) + (budget / *std::min_element(scaling_cost.begin(), scaling_cost.end())) * left;;  // A rough upper bound

    int max_throughput = left;  

    while (left <= right) {
        int mid = left + (right - left) / 2;
        
        if (canAchieveThroughput(throughput, scaling_cost, budget, mid)) {
            max_throughput = mid;  
            left = mid + 1;        
        } else {
            right = mid - 1;       
        }
    }

    return max_throughput;
}

int main() {
    vector<int> throughput = {4, 2, 7};
    vector<int> scaling_cost = {3, 5, 6};
    int budget = 32;

    int result = maxPipelineThroughput(throughput, scaling_cost, budget);
    cout << "Maximum achievable throughput: " << result << endl;

    return 0;
}
