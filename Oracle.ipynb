```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/



vector<int>parent;
vector<int>sizes;



int find(int node){
	if(parent[node]==node){
		return node;
	}

	return parent[node]=find(parent[node]);
}



bool unite(int a,int b){
	int x=find(a);
	int y=find(b);

	if(x==y)return false;
	if(sizes[x]<sizes[y]){
		parent[x]=y;
		sizes[y]+=sizes[x];
	}else{
		parent[y]=x;
		sizes[x]+=sizes[y];
	}

	return true;
}
void solve(){


	int n=4;
	// cin>>n;

	parent.assign(n+1,0);
	for(int i=0;i<=n;i++){
		parent[i]=i;
	}

	sizes.assign(n+1,1);

	vector<string> queryType = {"Friend", "Friend", "Total"};
    vector<int> student1 = {1, 2, 1};
    vector<int> student2 = {2, 3, 4};

    for(int i=0;i<queryType.size();i++){
    	if(queryType[i]=="Friend"){
    		unite(student1[i],student2[i]);
    	}else if(queryType[i]=="Total"){
    		if(parent[student1[i]]==parent[student2[i]]){
    			cout<<sizes[student1[i]]<<endl;
    		}else{
    			cout<<sizes[student1[i]]+sizes[student2[i]]<<endl;
    		}
    		

    	}
    }



}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}


```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){
    int depth=0;
    int ans=0;
    string s;
    cin>>s;

    for(int i=0;i<s.size();i++){
    	if(s[i]=='('){
    		depth++;
    	}else{
    		depth--;
    	}

    	if(depth==-1){
    		ans++;
    		depth++;
    	}
    }


    cout<<depth+ans<<endl;


    // stack<int>st;
    // int ans=0;

    // for(int i=0;i<s.size();i++){
    // 	if(s[i]=='('){
    // 		st.push(i);
    // 	}else{
    // 		if(!st.empty()){
    // 			st.pop();
    // 		}else{
    // 			ans++;
    // 		}
    // 	}
    // }

    // ans+=st.size();
    // cout<<ans<<endl;

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}





```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){


	int n;
	cin>>n;
	vector<int>arr(n);
	for(int i=0;i<n;i++){
		cin>>arr[i];
	}
	priority_queue<int,vector<int>,greater<int>>pq;

	for(int i=0;i<n;i++){
		pq.push(arr[i]);
	}

	int sum=0;
	while(pq.size()>1){
		int a=pq.top();
		pq.pop();
		int b=pq.top();
		pq.pop();
		sum+=(a+b);
		pq.push(a+b);


	}
	cout<<sum<<endl;

    

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}


```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){


	int n;
	cin>>n;
	vector<int>arr(n);
	for(int i=0;i<n;i++)cin>>arr[i];
	int m;
	cin>>m;
	vector<int>brr(m);
	for(int i=0;i<m;i++)cin>>brr[i];



	int i=0,j=0,k=0;
	vector<int>ans;
	while(i<n and j<m){
		if(arr[i]<brr[j]){
			ans.push_back(arr[i]);
			i++;
		}else{
			ans.push_back(brr[j]);
			j++;
		}
	}


	while(i<n){
		ans.push_back(arr[i]);
		i++;
	}


	while(j<m){
		ans.push_back(brr[j]);
		j++;
	}

	for(int i=0;i<ans.size();i++){
		cout<<ans[i]<<" ";
	}
	cout<<endl;



 

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}

```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){


	int n;
	cin>>n;
	vector<int>arr(n);
	for(int i=0;i<n;i++)cin>>arr[i];
	



	

	sort(arr.begin(),arr.end());

	int sum=0;
	for(int i=0;i<n;i++){
		if(i+1<n){
			sum+=(arr[i+1]-arr[i]);
		}

	}
	cout<<sum<<endl;



 

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}


```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){


	int x;
	cin>>x;

	if(x<0){
		cout<<"false"<<endl;
		return;
	}
	if(x==0){
		cout<<"true"<<endl;
		return;
	}

	if(!(x&(x-1))){
		cout<<"true"<<endl;
	}else{
		cout<<"false"<<endl;
	}

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}

```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/

void solve(){

	int n,m;
	cin>>n>>m;
	vector<vector<int>>adj(n+1);
	for(int i=0;i<m;i++){
		int u,v;
		cin>>u>>v;
		adj[u].push_back(v);
		adj[v].push_back(u);
	}
	vector<int>arr;

	vector<bool>visited(n+1,false);
	priority_queue<int>pq; 
	pq.push(n); 

	visited[n]=true;


	while(!pq.empty()){
		int node=pq.top();
		pq.pop();
		arr.push_back(node);

		for(auto it:adj[node]){
			if(!visited[it]){
				visited[it]=true;
				pq.push(it);
			}
		}
	}

	for(auto it:arr){
		cout<<it<<" ";
	} 
	cout<<endl;

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}

```javascript
async function getDiscountedPrice(barcode) {
    const url = `https://jsonmock.hackerrank.com/api/inventory?barcode=${barcode}`;

    try{

      const response=await fetch(url);
      const data= await response.json();

      if(data.length===0 || !data.data){
        return -1;
      }
      const item=data.data[0];
      const price=item.price;
      const discount=item.discount;
      const discountedprice=Math.round(price-((discount/100)*price))
      return discountedprice;

    }catch(error){
      console.error("Error",error)
    }
}

// Example usage:
getDiscountedPrice("74002314").then((result) => console.log(result));


```cpp
#include <iostream>
#include <vector>
#include <algorithm>

// Function to count the number of 1's in the binary representation of a number
int binaryCardinality(int num) {
    int count = 0;
    while (num > 0) {
        count += (num & 1); // Add 1 if the least significant bit is 1
        num >>= 1;          // Shift right to process the next bit
    }
    return count;
}

// Custom comparator function for sorting
bool compare(int a, int b) {
    int cardinalityA = binaryCardinality(a);
    int cardinalityB = binaryCardinality(b);

    // First sort by binary cardinality, then by decimal value
    if (cardinalityA == cardinalityB) {
        return a < b;
    }
    return cardinalityA < cardinalityB;
}

std::vector<int> sortByBinaryCardinality(std::vector<int>& arr) {
    std::sort(arr.begin(), arr.end(), compare);
    return arr;
}

int main() {
    std::vector<int> arr = {5, 3, 7, 10, 20};
    arr = sortByBinaryCardinality(arr);

    // Print the sorted array
    for (int num : arr) {
        std::cout << num << " ";
    }
    return 0;
}
