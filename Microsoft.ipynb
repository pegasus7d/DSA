```cpp
class Solution {
public:



    bool f(vector<int>&arr,int mid,long long  k){


        const int MAX_VAL = 100100;
        vector<int> freq(MAX_VAL+1);
        int head=-1,tail=0;
        int cnt=0;
        long long ans=0;
        while(tail<arr.size()){
            while(head+1<arr.size() && (cnt<mid || freq[arr[head+1]]>0)){
                head++;
                if(freq[arr[head]]==0){
                    cnt++;
                }
                freq[arr[head]]++;

            }

            ans+=(head-tail+1);
            if(tail>head){
                tail++;
                head=tail-1;
            }else{
                freq[arr[tail]]--;
                if(freq[arr[tail]]==0){
                    cnt--;
                }
                tail++;


            }
        }

        return ans>=k;

    }
    int medianOfUniquenessArray(vector<int>& nums) {
        int n=nums.size();
        set<int>st;
        for(auto &num:nums){
            st.insert(num);
        }

        int l=1,r=st.size();
        int ans=-1;
        long long total=n*1ll*(n+1)/2;
        long long median=(total+1)/2;


        while(l<=r){
            int mid=l+(r-l)/2;


            if(f(nums,mid,median)){
                ans=mid;
                r=mid-1;
            }else{
                l=mid+1;
            }
        }

        return ans;
        
    }
};
