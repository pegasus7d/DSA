```cpp
class Solution {
public:



    bool f(vector<int>&arr,int mid,long long  k){


        const int MAX_VAL = 100100;
        vector<int> freq(MAX_VAL+1);
        int head=-1,tail=0;
        int cnt=0;
        long long ans=0;
        while(tail<arr.size()){
            while(head+1<arr.size() && (cnt<mid || freq[arr[head+1]]>0)){
                head++;
                if(freq[arr[head]]==0){
                    cnt++;
                }
                freq[arr[head]]++;

            }

            ans+=(head-tail+1);
            if(tail>head){
                tail++;
                head=tail-1;
            }else{
                freq[arr[tail]]--;
                if(freq[arr[tail]]==0){
                    cnt--;
                }
                tail++;


            }
        }

        return ans>=k;

    }
    int medianOfUniquenessArray(vector<int>& nums) {
        int n=nums.size();
        set<int>st;
        for(auto &num:nums){
            st.insert(num);
        }

        int l=1,r=st.size();
        int ans=-1;
        long long total=n*1ll*(n+1)/2;
        long long median=(total+1)/2;


        while(l<=r){
            int mid=l+(r-l)/2;


            if(f(nums,mid,median)){
                ans=mid;
                r=mid-1;
            }else{
                l=mid+1;
            }
        }

        return ans;
        
    }
};

# Check Path Presence


```cpp


class Solution {
public:
    long long maximumImportance(int n, vector<vector<int>>& roads) {



        vector<int>degree(n);
        for(auto &it:roads){
            int u=it[0];
            int v=it[1];
            degree[u]++;
            degree[v]++;
        }


        sort(degree.begin(),degree.end());

        long long value=1;
        long long sum=0;
        for(int i=0;i<n;i++){
            sum+=degree[i]*value;
            value++;
        }
        return sum;
        
    }
};


Question 2:


There are two candy shops, both of which have N candies for sale. The first shop sells each candy for one golden coin; the second shop sells each candy for one silver coin. You have N/2 golden coins and N/2 silver ones (N is even).


Each candy is of some particular type (not necessarily unique), represented by an integer. What is the maximum number of different types of candy that you can buy?

```cpp
class Solution {
public:
    int maximumSetSize(vector<int>& nums1, vector<int>& nums2) {
        int n=nums1.size();
        unordered_set<int>s1,s2,common;
        for(auto &x:nums1){
            s1.insert(x);
        }
        for(auto &x:nums2){
            s2.insert(x);
            if(s1.find(x)!=s1.end()){
                common.insert(x);
            }
        }

       int n1=s1.size(),n2=s2.size(),c=common.size();
        return min(min(n/2,n1-c)+min(n/2,n2-c)+c,n);
    }
};

# Find Suitable Locations


```cpp
#include <bits/stdc++.h> 
int findBestMeetingPoint(vector<vector<int>> &mat)
{
     vector<int>x,y;
     vector<pair<int,int>>pts;
     int n=mat.size();
     int m=mat[0].size();
     for(int i=0;i<n;i++){
          for(int j=0;j<m;j++){
               if(mat[i][j]){
                    x.push_back(i);
                    y.push_back(j);
                    pts.push_back(make_pair(i,j));
               }
               
          }
     }



     sort(x.begin(),x.end());
     sort(y.begin(),y.end());
     int a,b;
     int ans=0;

     if(x.size()&1){
          a=x[x.size()/2];
          b=y[y.size()/2];

     }else{
          a=x[x.size()/2-1];
          b=y[y.size()/2-1];
     }

     for(auto it:pts){
          ans+=(abs(a-it.first));
          ans+=(abs(b-it.second));
     }
     return ans;
}



```cpp
int f(vector<int>&arr,int i,int x,int y){
    if(i==arr.size()){
        if(x>0 && y>0){
            if(x%y==0 || y%x==0){
                return 1;
            }
        }
        return 0;
    }



    string key=to_string(i)+"_"+to_string(x)+"_"+to_string(y);
    if(dp.find(key)!=dp.end()){
        return dp[key];
    }


    long long ans=0;

    ans=(ans+f(arr,i+1,x,y))%mod;
    ans=ans+f(arr,i+1,x^arr[i],y)%mod;
    ans=ans+f(arr,i+1,x,y^arr[i])%mod;


    return dp[key]=ans;
}

```cpp
class Solution {
public:
    void dfs(int node,int parent,vector<vector<pair<int,int>>>&adj,int &ans){


        for(auto it:adj[node]){
            if(it.first!=parent){
                if(it.second==1){
                    ans++;
                }
                dfs(it.first,node,adj,ans);
            }
        }





    }
    int minReorder(int n, vector<vector<int>>& connections) {


        vector<vector<pair<int,int>>>adj(n+1);
        for(auto it:connections){
            int u=it[0];
            int v=it[1];
            adj[u].push_back({v,1});
            adj[v].push_back({u,0});
        }
        int ans=0;

        dfs(0,-1,adj,ans);
        return ans;


        

        
    }
};


```cpp
class Solution {
public:
    int dfs(vector<vector<int>>&adj,int node,int par,string &s,int &result){
        int a=0;
        int b=0;

        for(auto it:adj[node]){
            if(it!=par){


                int c=dfs(adj,it,node,s,result);
                if(s[it]==s[node])continue;


                if(c>b)b=c;
                if(b>a)swap(a,b);

            }
        }


        int case1=max(a,b)+1;
        int case2=1;
        int case3=1+a+b;//converges


        result=max({result,case1,case2,case3});


        return max(case1,case2);

    }
    int longestPath(vector<int>& parent, string s) {

        vector<vector<int>>adj(parent.size()+1);

        for(int i=1;i<parent.size();i++){
            int u=i;
            int v=parent[i];
            adj[u].push_back(v);
            adj[v].push_back(u);
        }

        int result=0;
        dfs(adj,0,-1,s,result);

        return result;





        
    }
};

```cpp
#include <bits/stdc++.h>
using namespace std;
#define endl '\n'



/*
#include <ext/pb_ds/assoc_container.hpp>
#include <ext/pb_ds/tree_policy.hpp>
using namespace __gnu_pbds;

typedef tree<int, null_type, less_equal<int>, rb_tree_tag, tree_order_statistics_node_update> ordered_set;

*/

#define int long long
/*

os.find_by_order(os.order_of_key(num));      
os.order_of_key(num) position of number 
*/


vector<int>dp;

int f(int i,vector<int>&arr){
	if(i>=arr.size())return 0;
	if(dp[i]!=-1)return dp[i];

	int ans=arr[i]+f(i+1,arr);

	


	if(i+1<arr.size()){
		string s1=to_string(arr[i]);
		string s2=to_string(arr[i+1]);
		int merge=stoll(s1+s2);
		ans=max(ans,merge+f(i+2,arr));
	}
	return dp[i]=ans;

}




void solve(){

	int n;
    cin>>n;
    vector<int>arr={12, 6, 18, 10, 1, 0};
    for(int i=0;i<n;i++){
        cin>>arr[i];
    }

    dp.assign(100100,-1);


    cout<<f(0,arr);
    

    

}

signed main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);cout.tie(NULL);	
    // int _t;cin>>_t;while(_t--)
    solve();
}

```cpp
int dp[101][27][101][101];

class Solution {
public:

    int f(string &s,int i,int prev,int freq,int k){
        if(k<0){
            return INT_MAX;
        }

        if(i>=s.size()){
            return 0;
        }

        if(dp[i][prev][freq][k]!=-1)return dp[i][prev][freq][k];

        int delete_i=f(s,i+1,prev,freq,k-1);
        int keep_i=0;

        if(s[i]-'a'!=prev){
            keep_i=1+f(s,i+1,s[i]-'a',1,k);
        }else{

            int one_more_addition=0;
            if(freq==1 || freq==9 || freq==99){
                one_more_addition=1;
            }
            keep_i=one_more_addition+f(s,i+1,prev,freq+1,k);
        }

        return dp[i][prev][freq][k]=min(delete_i,keep_i);



    }
    int getLengthOfOptimalCompression(string s, int k) {

        memset(dp,-1,sizeof dp);
        return f(s,0,26,0,k);



        
    }
};
