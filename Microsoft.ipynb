```cpp
class Solution {
public:



    bool f(vector<int>&arr,int mid,long long  k){


        const int MAX_VAL = 100100;
        vector<int> freq(MAX_VAL+1);
        int head=-1,tail=0;
        int cnt=0;
        long long ans=0;
        while(tail<arr.size()){
            while(head+1<arr.size() && (cnt<mid || freq[arr[head+1]]>0)){
                head++;
                if(freq[arr[head]]==0){
                    cnt++;
                }
                freq[arr[head]]++;

            }

            ans+=(head-tail+1);
            if(tail>head){
                tail++;
                head=tail-1;
            }else{
                freq[arr[tail]]--;
                if(freq[arr[tail]]==0){
                    cnt--;
                }
                tail++;


            }
        }

        return ans>=k;

    }
    int medianOfUniquenessArray(vector<int>& nums) {
        int n=nums.size();
        set<int>st;
        for(auto &num:nums){
            st.insert(num);
        }

        int l=1,r=st.size();
        int ans=-1;
        long long total=n*1ll*(n+1)/2;
        long long median=(total+1)/2;


        while(l<=r){
            int mid=l+(r-l)/2;


            if(f(nums,mid,median)){
                ans=mid;
                r=mid-1;
            }else{
                l=mid+1;
            }
        }

        return ans;
        
    }
};

# Check Path Presence


```cpp


class Solution {
public:
    long long maximumImportance(int n, vector<vector<int>>& roads) {



        vector<int>degree(n);
        for(auto &it:roads){
            int u=it[0];
            int v=it[1];
            degree[u]++;
            degree[v]++;
        }


        sort(degree.begin(),degree.end());

        long long value=1;
        long long sum=0;
        for(int i=0;i<n;i++){
            sum+=degree[i]*value;
            value++;
        }
        return sum;
        
    }
};


Question 2:


There are two candy shops, both of which have N candies for sale. The first shop sells each candy for one golden coin; the second shop sells each candy for one silver coin. You have N/2 golden coins and N/2 silver ones (N is even).


Each candy is of some particular type (not necessarily unique), represented by an integer. What is the maximum number of different types of candy that you can buy?

```cpp
class Solution {
public:
    int maximumSetSize(vector<int>& nums1, vector<int>& nums2) {
        int n=nums1.size();
        unordered_set<int>s1,s2,common;
        for(auto &x:nums1){
            s1.insert(x);
        }
        for(auto &x:nums2){
            s2.insert(x);
            if(s1.find(x)!=s1.end()){
                common.insert(x);
            }
        }

       int n1=s1.size(),n2=s2.size(),c=common.size();
        return min(min(n/2,n1-c)+min(n/2,n2-c)+c,n);
    }
};

# Find Suitable Locations


```cpp
#include <bits/stdc++.h> 
int findBestMeetingPoint(vector<vector<int>> &mat)
{
     vector<int>x,y;
     vector<pair<int,int>>pts;
     int n=mat.size();
     int m=mat[0].size();
     for(int i=0;i<n;i++){
          for(int j=0;j<m;j++){
               if(mat[i][j]){
                    x.push_back(i);
                    y.push_back(j);
                    pts.push_back(make_pair(i,j));
               }
               
          }
     }



     sort(x.begin(),x.end());
     sort(y.begin(),y.end());
     int a,b;
     int ans=0;

     if(x.size()&1){
          a=x[x.size()/2];
          b=y[y.size()/2];

     }else{
          a=x[x.size()/2-1];
          b=y[y.size()/2-1];
     }

     for(auto it:pts){
          ans+=(abs(a-it.first));
          ans+=(abs(b-it.second));
     }
     return ans;
}



```cpp
int f(vector<int>&arr,int i,int x,int y){
    if(i==arr.size()){
        if(x>0 && y>0){
            if(x%y==0 || y%x==0){
                return 1;
            }
        }
        return 0;
    }



    string key=to_string(i)+"_"+to_string(x)+"_"+to_string(y);
    if(dp.find(key)!=dp.end()){
        return dp[key];
    }


    long long ans=0;

    ans=(ans+f(arr,i+1,x,y))%mod;
    ans=ans+f(arr,i+1,x^arr[i],y)%mod;
    ans=ans+f(arr,i+1,x,y^arr[i])%mod;


    return dp[key]=ans;
}
